```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
sns.set(font_scale=1.3)
wine = pd.read_csv('winequality-red.csv')
print(wine)
```

```python
import seaborn as sns
wine.isnull().sum()
```

```python
sns.pairplot(wine);
```

```python
sns.barplot(x=wine.quality, y=wine.alcohol)
```

```python
sns.countplot(y='quality',data=wine,palette='colorblind')
plt.xlabel('')
plt.ylabel('')
plt.show()
```

```python
sns.set_style('whitegrid')
plt.title('    ',size=16)
plt.xlabel('residual sugar',size=12)
plt.ylabel('alcohol',size=12)
sns.scatterplot(x='residual sugar', y='alcohol', data=wine, hue='quality', s=90, edgecolor='black', palette='colorblind',hue_order=[3,4,5,6,7,8])
plt.show()
```

```python
plt.boxplot(wine['fixed acidity'])
plt.ylabel('fixed acidity')
plt.title('   ')
plt.show()
```
